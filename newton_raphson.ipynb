# Newton Fractal

Boundary set for:
$$
z^3 - 2z + 2 = 0
$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# coefficients for z^3 - 2z + 2 = 0
coeffs = [1, 0, -2, 2]

roots = np.roots(coeffs)
r_1, r_2, r_3 = roots

def converges_to(z):
    '''
    Takes a point z in the complex plane and returns the corresponding colour of convergence
    Returns black if the point diverges.
    '''
    
    delta = 0.01 # convergence radius - if the point gets within delta of the root it converges to that root.
   
    for x in range(60):
        f_z = (z**3) - 2 * z + 2 # polynomial
        f_z_der = 3 * (z**2) - 2 # it's derivative
        z -= f_z / f_z_der # Newton - Raphson formula
        
        if abs(z - r_3) < delta:
            for i in range(20):
                if x == i:
                    return (1 - i*0.05, 0 , 0)
            
        if abs(z - r_2) < delta:
            for i in range(20):
                if x == i:
                    return (1 - 0.05 * i, 1 - 0.05 * i, 1 - 0.05 * i)
           
        if abs(z - r_1) < delta:
            for i in range(20):
                if x == i:
                    return (0, 0, 1 - 0.05 * i)
    return 'black'  #(diverged)

x = []
y = []
colour = []
for a in range(-400, 401):
    for b in range(-400, 401):
        z = a/200 + b/200 * 1j
        x.append(a/200)
        y.append(b/200)
        colour.append(converges_to(z))

plt.figure(figsize=(10,10))
plt.scatter(x, y, c=colour, s=1)
plt.show()
```
