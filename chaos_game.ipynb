# Chaos Game

## Introduction

The Chaos Game is a mathematical algorithm that generates fractal patterns through simple iterative rules. The classic version uses a triangle, but this implementation extends it to a regular pentagon.

The algorithm works by:
1. Starting with a regular pentagon defined by 5 vertices (A, B, C, D, E)
2. Beginning at a random point inside the pentagon
3. Repeatedly choosing a random vertex and moving halfway toward it
4. Plotting each new point to reveal the fractal pattern

## Implementation

This implementation uses Python with Matplotlib for visualization. The code consists of three main components:

1. **Pentagon Generation**: Calculating the coordinates of a regular pentagon's vertices using trigonometric functions
2. **Chaos Game Simulation**: Iterative point generation following the chaos game rules
3. **Visualization**: Scatter plot rendering of the resulting fractal pattern

Key parameters:
- **FACTOR**: Division factor (2 means halfway point but can be changed for different results)
- **Iterations**: 120,000 points for detailed fractal generation
- **Starting point**: (0.5, 0.5) as initial position

The algorithm ensures no consecutive moves to the same vertex.

```python
import random, math
import matplotlib.pyplot as plt
```

```python
points = []
for i in range(5):
    point = [math.cos((math.pi * (90 + 72*i))/ 180), math.sin((math.pi * (90 + 72*i)) / 180)]
    points.append(point)
A, B, C, D, E = points
X = [0.5, 0.5]

x_to_plot, y_to_plot = [], []
FACTOR = 2
current = X
for _ in range(120000):
    r = random.randint(0, 4)
    if r == current:
        continue
    else:
        current = r
    xn = (points[r][0] + X[0]) / FACTOR
    yn = (points[r][1] + X[1]) / FACTOR
    x_to_plot.append(xn)
    y_to_plot.append(yn)
    X = (xn, yn)
plt.figure(figsize=(12, 12))
plt.scatter(x_to_plot, y_to_plot, s=1);
```

## Result

The visualization above shows the fractal pattern generated by the chaos game on a pentagon.

Key observations from the pattern:
- Points cluster densely in certain regions while leaving others relatively sparse
- The structure shows self-similarity at different scales
- The boundary regions between vertices show intricate detail
